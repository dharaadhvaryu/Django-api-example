```python
# prompt: print which embark town has highest mean fare

def highest_mean_fare(df):
  embark_town_fares = df.groupby('embark_town')['fare'].mean()
  highest_mean_fare = embark_town_fares.max()
  highest_mean_fare_town = embark_town_fares.idxmax()
  return highest_mean_fare_town, highest_mean_fare

highest_mean_fare_town, highest_mean_fare = highest_mean_fare(df)
print(f"Embark town with highest mean fare: {highest_mean_fare_town}")
print(f"Highest mean fare: {highest_mean_fare}")
```
